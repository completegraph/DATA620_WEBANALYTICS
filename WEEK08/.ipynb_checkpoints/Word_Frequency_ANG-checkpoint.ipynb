## A. Selecting a Corpus

We choose the Brown corpus because it was the first major linguistic corpus of American Standard English available for computational study .   The Brown corpus contained 500 selected texts written in 1961 by native speakers of American English.  Each selected text contained approximately 2000 words and was taken from multiple genres.  The Brown corpus was compiled by linguists Henry Kucera and W. Nelson Francis (https://en.wikipedia.org/wiki/Brown_Corpus).

The Brown corpus is available in the `nltk` package and is highlighted in its documentation.   In the following, we import the Brown corpus and obtain basic statistics about its unique words.   An initial subset of the file identifiers of the corpus is shown below.


```python
import pandas as pd
import nltk
from nltk.corpus import brown
print(brown.fileids()[0:30])
```

The following code snippet examines size and shows the first 500 characters of the first text sample.   We see there are 2242 words in the first file.

```python
ca01 = brown.words('ca01')
print(len(ca01))
print(" ".join(ca01)[0:500])
```

The Brown corpus is divided into categories.  We count the words in each category and aggregate the total for our first word count.

```python
brown_categories = brown.categories()

total_words = 0
print("Word Count by category and for the entire corpus\n------------------------------")
for cat in brown_categories:
    text_cat = brown.words( categories = cat )
    print(cat, "  ",  len(text_cat) )
    total_words += len(text_cat)
    
print("-----------------------------\nFinal count : ", total_words)
```

While the corpus has 1,161,192 words, that count includes punctuations, as we will show below, which we would not really consider words.  

The following data transformations handle 3 issues:

*  Punctuation are excluded from word counts

*  Words without any alphabetical characters `[a-zA-Z]` are excluded.  This happens for strings like `''`.

*  Upper and lower case words are simplified to being equivalent.

As a result we obtain just over 1.005 million words in the Brown corpus.


```python
brown_words = brown.words()

from string import punctuation
print(punctuation)

brown_clean_1 = [ u for u in brown_words if u not in punctuation ]

brown_clean_2 = [ u for u in brown_clean_1 if ( any( c.isalpha() for c in u ) ) ]

brown_clean_3 = [ u.lower() for u in brown_clean_2 ]

brown_without_punct = brown_clean_3

brown_N = len(brown_without_punct)

print(brown_N , " words after cleaning up these 3 issues.")
```

## B. Frequency Distribution of Most Common Words

The frequency distribution of the most common words is computed without stemming and lemmatization.
We obtain the `nltk.FreqDist` frequency distribution object from the Brown corpus and find 48,052 distinct words.
Then we compute the subset of 200 most common words using the `most_common` function and report their relative frequency, word count in descending order.

```python
brown_vocab = nltk.FreqDist(brown_without_punct)
brown_vocab_N = len(brown_vocab)
brown_200 = brown_vocab.most_common(200)

print(brown_vocab_N , " total distinct words from the cleaned word list." )
```

```python
df_word_freq = pd.DataFrame(columns = ['word', 'count', 'freq' , 'cumfreq', 'rank' ] )
    
cum_count = 0
num_words = 0
for tup in brown_200:
        num_words = num_words + 1
        cum_count = cum_count + tup[1]
        
        df_word_freq = df_word_freq.append( { 'word' : tup[0], 
                                             'count' : tup[1] , 
                                             'freq': tup[1]/brown_N ,
                                             'cumfreq': cum_count / brown_N ,
                                             'rank' : num_words }, ignore_index = True )
df_word_freq
```

The table shows:

*   `the`, `of`, `and` are the most frequently occurring words.
*   the first 200 most common words represent about 54% of all words.  The 200th word is `took`.
*   The 127th word has cumulative frequency 50.0108%.  So that means 127 words represent just over half the corpus as shown below.


```python
df_word_freq[ df_word_freq["cumfreq"] > 0.5 ]
```

## C. Visualizing the Frequency Data

The visualization of the top 200 words is impractical in a single graph.


```python
import numpy as np
import seaborn as sns
```

```python
def plot_freq_range(i):
    
    num_words = 50
    sns.set(rc = {'figure.figsize':(8,12)}, font_scale = 1.5)
    ax = sns.barplot(data = df_word_freq[i*num_words :(i+1)*num_words] 
            ,x = 'freq'
            ,y = 'word'
            )

    ax.set(title="Brown Corpus Word Frequencies {}-{}".format( i*num_words+1, (i+1)*num_words ) )
    ax.bar_label(ax.containers[0])

```

```python
plot_freq_range(0)
```

```python
plot_freq_range(1)
```

```python
plot_freq_range(2)
```

```python
plot_freq_range(3)
```

## D. Zipf Distribution and All Corpora

We can check if the Zipf distribution law holds for the Brown Corpus by plotting the log transformed values of the rank and frequency of the top 200 words.


```python
import math
df_word_freq["logrank"] = [ math.log(w) for w in  df_word_freq["rank"] ]
df_word_freq["logfreq"] = [ math.log(w) for w in  df_word_freq["freq"] ]

m, b = np.polyfit( df_word_freq["logrank"], df_word_freq["logfreq"], 1 )

print("m ", m, " b", b )
    
```

```python
math.exp(b)
```

We conclude that an empirical model fit based on the linear regression of the log transformed data gives:

$$log(\text{Freq}) = -2.186 - 1.04 \cdot log(\text{Rank})$$

which is equivalent to:

$$\text{Freq} = \frac{0.112}{\text{Rank}^{1.04}}$$

This is a reasonable fit for the top 200 words to Zipf's law which predicts:

$$\text{Freq} = \frac{K}{\text{Rank}^1} \text{ for some constant } K$$

Lastly, we visualize the resulting log-log regression fit below.

```python
f, ax = plt.subplots(figsize=(12, 12))
ax.set(title="Brown Corpus Word Frequencies Zipf Law Comparison" )
sns.regplot(x="logrank", y="logfreq", data=df_word_freq , ax=ax)

plt.text(2, -3, "log(Freq) = -2.185 - 1.04 * log(Rank)", 
         horizontalalignment='left', size='medium', color='red', weight='semibold')
```

Zipf's Law is remarkably accurate for words with ranks between 5-200.  For the most common 3 words, the predicted frequency is somewhat higher than the observed frequency.

We speculate that the Brown Corpus varies somewhat from other corpora in their usage of words.   
Because the Brown Corpus was based on text gathered in 1961, the usage of words has changed over time.  For example, existing words like `mouse` have gained meanings as a computer device not just a rodent.  While other words like `email`, `internet` did not exist in 1961.

Despite changes in usage over time, we speculate that modern day corpora likewise conform to Zipf's law.  This is supported by the current understanding of word frequency in all languages, although the reason for this is not well understood.  https://en.wikipedia.org/wiki/Zipf%27s_law
