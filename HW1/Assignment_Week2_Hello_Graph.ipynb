In order to use networkx 1.11 and plot the graphs, we installed an order version of `matplotlib` reverting to version 2.2.3 as well.

To implement the graph, we will define a graph $G$ using numeric nodes.
Then we assign labels in a dictionary and attach the dictionary to the nodes in $G$.

```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G=nx.Graph()   # Creates an empty graph
```

```python
G.add_nodes_from(range(10))  # Creates 10 integer nodes from 0..9
```

```python
Node_names = dict( zip( range(10),     ['Andre', 'Beverly', 'Carol', 'Diane', 'Ed', 'Fernando', 'Garth', 'Heather', 'Ike', 'Jane']) )
    
```

```python
Node_names
```

Let's inspect the graph so far.   We first show the graph with no edges.  Note that the nodes are numbers from 0 to 9 but the labels are specified above using `zip` to align the node integers to names.

```python
nx.draw(G, labels = Node_names, with_labels = True , alpha = 0.5, font_size = 16 , font_color = 'red',  node_color = 'gray' )
```

Next, we specify the edges using `add_edges_from` in the form of a list of pairs.

```python
edge_list = [ ( 0,1),  (0, 2),  (0, 3), (0, 5), 
              (1, 3 ), (1, 4), (1, 6) ,
              (2, 3), (2, 5 ) ,
              (3, 4), (3, 5 ) , (3, 6 ), 
              (4, 6), 
              (5, 6), (5, 7 ),
              (6, 7 ), 
              (7, 8 ), 
              (8, 9 )
            ]
```

```python
G.add_edges_from(edge_list )
```

When we draw the graph with the included edges, we add some formatting parameters.

* `labels` specify the list of node names and is required for our use case
* `with_labels` needs to be set to `True` when using `labels` parameter and is required for our use case
* `alpha` controls the transparency of nodes and edges and is optional
* `font_size` and `font_color` and `node_color` are optional 
* `width` allows us to control edge thickness, but can be a scalar or list.  We use a scalar since all edges have the same width.

```python
nx.draw(G, labels = Node_names, with_labels = True , alpha = 0.5, font_size = 16 , font_color = 'red', 
        node_color = 'gray', width = 5 )
```

One unfortunate consequence of the above drawing is that graph node positions are more or less random.  They don't look like the required shape.   Using an approximate integer grid system starting on the bottom left corner of the graph, we lay out the positions for all the nodes in a dictionary called `pos` and then add an optional parameter `pos` to the `draw` function.  We also add an `edge_color` parameter to more closely match the stated design.


```python
pos = { 0: (0, 4), 1:(0, 1), 
            2:( 1, 5), 3:(1, 3), 
             4:(1,0), 5:(2, 4), 
             6:(2, 1), 7: (3,3), 
             8:(4, 3), 9:(5, 3) }
```

The graph output below represents our complete functional specification of the network.
We decide not to include custom icons for the nodes.

```python

nx.draw(G, pos = pos, labels = Node_names, with_labels = True , alpha = 0.5, font_size = 16 , font_color = 'red', 
        node_color = 'gray' , edge_color = 'purple' , width = 5 )


# Save the figure to file.  Must be in the same cell otherwise, the figure is blank when saved.
plt.savefig("Assignment1_Week2_Purple_Graph.png")
```

We do consider the effort to replicate the homework with a border around the graph.  This seems to be a bit of a hack to add the border, but works.

```python
fig = plt.figure(linewidth = 0.5)
# Remove tick marks around the figure border
ax = plt.gca()
ax.axes.xaxis.set_ticks([])
ax.axes.yaxis.set_ticks([])

nx.draw(G, pos = pos, labels = Node_names, with_labels = True , alpha = 0.5, font_size = 16 , font_color = 'red', 
        node_color = 'gray' , edge_color = 'purple' , width = 5 )

# Add the purple border around the plot (optional)
fig.set_facecolor("purple")
plt.axis("on")

# Save the figure to file.  Must be in the same cell otherwise, the figure is blank when saved.
plt.savefig("Assignment1_Week2_Purple_Graph2.png")
```
